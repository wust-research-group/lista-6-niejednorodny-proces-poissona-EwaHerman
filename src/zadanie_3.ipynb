```python
import numpy as np
import matplotlib.pyplot as plt

def lambda_function1(t):
    return np.log(t)
    
def lambda_function2(t):   
    return 5*t**(1/3)

def dodane_lambda(t):
    return np.log(t) + 5*t**(1/3)

T=30
l1=np.log(T)
l2=5*T**(1/3)
l=l1+l2

def polaczone(T,lambda_function1,lambda_function2,l1,l2):
    S1, I1 = niejednorodny_poiss(T, lambda_function1, l1)
    S2, I2 = niejednorodny_poiss(T, lambda_function2, l2)
    S3 = S1 + S2
    S = sorted(S3)
    I = I1 + I2
    return S, I

def niejednorodny_poiss(T, lambda_function, l):
    t = 0
    I = 0
    S = []
    while t <= T:
        U1 = np.random.random()
        t -= np.log(U1) / l
        if t > T:
            break
        U2 = np.random.random()
        if U2 <= lambda_function(t) / l:
            I += 1
            S.append(t)
    return S, I

S1, I1 = polaczone(T,lambda_function1,lambda_function2,l1,l2)
S2, I2 =niejednorodny_poiss(T, dodane_lambda, l)

plt.scatter(S1, range(len(S1)))
plt.scatter(S2, range(len(S2)))
plt.xlabel('Czas (t)')
plt.ylabel('Liczba zdarzeń (N(t))')
plt.grid(True)
plt.show()
```
